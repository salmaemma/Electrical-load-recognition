```python
mask = np.array(rt.iloc[:, -6:7])
res = mask.astype(np.float)
np.set_printoptions(precision=3, suppress=True)
print(res)
print(len(res), res.shape)
```

```python
axis_label = {0:'voltage', 1:'Arus'}

#Frequency Sampling
fs = 50

for i in range (1,7):
    label_list = []
    header = 'v_har, A_har, volt, current, activepower, reactivepower, apparentpower'
    mask1 = res[:,4]==i
    
    for j in (axis_label) :
        fig, (ampere1, ampere2, ampere3) = plt.subplots(3,1)
        plt.subplots_adjust(hspace = 0.4)
        plt.figure(figsize=(10,8))
        fig.suptitle('label {}_{}'.format(i,axis_label[j]))
        alx = res[mask1,j]
    
        n = len(alx)
        cut_len = int(n*0.1)
        b = alx[cut_len:-cut_len]
        m = b.size
        
        #frequency domain
        freqY = np.fft.fft(b)
        freq = np.fft.fftfreq(m, d=1/fs)
        shift_freq = np.fft.fftshift(freq)
                
        #Spectrum
        spectrum = np.sqrt(freqY.real**2+freqY.imag**2)               
        shift_spec = np.fft.fftshift(spectrum)
        spec = np.fft.ifft(spectrum)
        
        #harmonic
        har=freqY/n
        har=har[range(n//10)]
        
        #label_list.append(b.reshape(-1,1))
        #label_list.append(spec.real.reshape(-1,1))
        label_list.append(har.real.reshape(-1,1))
        ampere1.set_ylim(0,100)
        ampere1.plot(shift_freq, shift_spec, label='{}_Spec. Harmonik'.format(axis_label[j]))
        legend = ampere1.legend(loc='best', shadow=True, fontsize='small')
      
        
        ampere2.set_ylim(0,0.05)
        ampere2.plot(har, label='{}_Gel. Harmonik'.format(axis_label[j]))
        legend = ampere2.legend(loc='best', shadow=True, fontsize='small')
        
        ampere3.plot(b, label='{}'.format(axis_label[j]))
        legend = ampere3.legend(loc='best', shadow=True, fontsize='small')
        
    voltage=res[mask1, 0]
    current=res[mask1, 1]
    power=res[mask1, 2]
    powerfactor=res[mask1, 3]
    reactp = voltage * current
    apparentp = reactp * powerfactor
    #label=res[mask1, 4]
        
    nrow = min(har.shape[0], b.shape[0])
    label_list.append(voltage[:nrow].reshape(-1,1))
    label_list.append(current[:nrow].reshape(-1,1))
    label_list.append(power[:nrow].reshape(-1,1))
    label_list.append(reactp[:nrow].reshape(-1,1))
    label_list.append(apparentp[:nrow].reshape(-1,1))
    #label_list.append(label[:nrow].reshape(-1,1))
        
    arrange_table = np.hstack(label_list)
    print(arrange_table.shape)
    np.savetxt(r'loaddata/label/label_{}.csv'.format(i), arrange_table, delimiter=',', fmt='%18f', header=header, comments='')
```
